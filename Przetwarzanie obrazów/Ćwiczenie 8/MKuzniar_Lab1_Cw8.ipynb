```python
import numpy as np 
import cv2 as cv
import matplotlib.pyplot as plt
from PIL import Image
```

Zbadaj skutecznosc redukcji szumu typu „sól i pieprz” za pomoca

a) liniowego filtra usredniajacego z kwadratowa maska, rozpoczynajac od maski
rozmiaru 3 × 3

<em> **cv.blur()** - In this process, the central element of the image is replaced by the average of 
all the pixels in the kernel area. 

```python
fname = 'US plane.png'
image = Image.open(fname).convert("L")
img = np.asarray(image)

plt.figure(figsize=(15,10))

plt.subplot(2,2,1)
plt.title("Obraz oryginalny")
plt.imshow(img, cmap='gray', vmin=0, vmax=255)

plt.subplot(2,2,2)
plt.title("Maska 3x3")
plt.imshow(cv.blur(img,(3,3)), cmap='gray', vmin=0, vmax=255)   #cv. blur() - maska 3x3

plt.subplot(2,2,3)
plt.title("Maska 7x7")
plt.imshow(cv.blur(img,(7,7)), cmap='gray', vmin=0, vmax=255)   #cv. blur() - maska 7x7

plt.subplot(2,2,4)
plt.title("Maska 15x15")
plt.imshow(cv.blur(img,(15,15)), cmap='gray', vmin=0, vmax=255) #cv. blur() - maska 15x15
```

b) nieliniowego filtra medianowego

<em> **cv.medianBlur()** - central element of the image is replaced by the median of all the pixels in the kernel area. This operation processes the edges while removing the noise.

```python
plt.figure(figsize=(15,10))

plt.subplot(2,2,1)
plt.title("Obraz oryginalny")
plt.imshow(img, cmap='gray', vmin=0, vmax=255)

plt.subplot(2,2,2)
plt.title("Maska 3x3")
plt.imshow(cv.medianBlur(img,3), cmap='gray', vmin=0, vmax=255) # cv.medianBlur() - maska 3x3

plt.subplot(2,2,3)
plt.title("Maska 7x7")
plt.imshow(cv.medianBlur(img,7), cmap='gray', vmin=0, vmax=255) # cv.medianBlur() - maska 7x7

plt.subplot(2,2,4)
plt.title("Maska 15x15")
plt.imshow(cv.medianBlur(img,15), cmap='gray', vmin=0, vmax=255) # cv.medianBlur() - maska 15X15
```

Zauważyć można, że użycie nieliniowego filtra medianowego dało lepsze rezutaty przy redukcji szumu typu sól i piperz. Subiektywnie oceniając najlepszą redukcję można zaobserwować przy użyciu nieliniowego filtra medianowego o masce 3x3. Największa redukcja szumów przy jednoczesnym zachowaniu szczegółów zdjęcia.
