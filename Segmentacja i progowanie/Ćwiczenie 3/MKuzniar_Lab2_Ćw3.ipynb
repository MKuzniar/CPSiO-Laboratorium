```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
from PIL import Image
from scipy import signal
from scipy import ndimage
```

```python
fname = 'micro.png'
image_grayscale = Image.open(fname).convert("L")
img_gray = np.asarray(image_grayscale)
```

## Globalne progowanie

```python
img_blur = cv.medianBlur(img_gray,5)

ret1,threshold = cv.threshold(img_blur,127,255,cv.THRESH_BINARY)

plt.figure(figsize=(17,12))

plt.subplot(1,3,1)
plt.title("Obraz oryginalny")
plt.imshow(img_gray, cmap='gray', vmin=0, vmax=255)

plt.subplot(1,3,2)
plt.title("Obraz wygładzony")
plt.imshow(img_blur, cmap='gray', vmin=0, vmax=255)

plt.subplot(1,3,3)
plt.title("Obraz posegmentowany")
plt.imshow(threshold, cmap='gray', vmin=0, vmax=255)
```

## Adaptacyjne progrowanie

```python
img_blur = cv.medianBlur(img_gray,5)

adapt_threshold = cv.adaptiveThreshold(img_blur,255,cv.ADAPTIVE_THRESH_GAUSSIAN_C, cv.THRESH_BINARY_INV,11,2)

plt.figure(figsize=(17,12))

plt.subplot(1,3,1)
plt.title("Obraz oryginalny")
plt.imshow(img_gray, cmap='gray', vmin=0, vmax=255)

plt.subplot(1,3,2)
plt.title("Obraz wygładzony")
plt.imshow(img_blur, cmap='gray', vmin=0, vmax=255)

plt.subplot(1,3,3)
plt.title("Obraz posegmentowany")
plt.imshow(adapt_threshold, cmap='gray', vmin=0, vmax=255)
```

## Optymalne progrowanie metodą Otsu

```python
img_blur = cv.medianBlur(img_gray,5)

ret2,threshold_otsu = cv.threshold(img_blur,0,255,cv.THRESH_BINARY+cv.THRESH_OTSU)

plt.figure(figsize=(17,12))

plt.subplot(1,3,1)
plt.title("Obraz oryginalny")
plt.imshow(img_gray, cmap='gray', vmin=0, vmax=255)

plt.subplot(1,3,2)
plt.title("Obraz wygładzony")
plt.imshow(img_blur, cmap='gray', vmin=0, vmax=255)

plt.subplot(1,3,3)
plt.title("Obraz posegmentowany")
plt.imshow(threshold_otsu, cmap='gray', vmin=0, vmax=255)
```
